```python
# pip install requests
# pip install boto3
# pip install base64
# pip install matplotlib

# https://huggingface.co/datasets
# https://teachablemachine.withgoogle.com/
# https://aws.amazon.com/ko/rekognition/

import boto3
import base64

rekognition = boto3.client(
    'rekognition',
    aws_access_key_id='/* access key */',
    aws_secret_access_key='/* secret key */',
    region_name='ap-northeast-2'
)

# with:       
# =>       .
# rb: read in binary mode
with open('./images/landscape.jpeg', 'rb') as cf:
    # AWS Rekognition   binary  
    base64_image = base64.b64encode(cf.read())
    base64_binary = base64.decodebytes(base64_image)
    print("file open")

response = rekognition.detect_labels(Image={'Bytes':base64_binary})
response
```
